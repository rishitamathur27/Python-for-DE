## Pivot Tables: Excellent way to summarize your data!

A pivot table is a tool that allows you to reorganize and summarize selected columns and rows of data in a dataframe

Pivot table provides an easy way to subset by  one column and then apply a calculation like a sum or a mean


Pivot table first groups and then applies a calculation

```python
data = {
    'A': ['a','b','b','a'],
    'B': ['p','q','r','s'],
    'C': ['x','y','x','y'],
    'D': [1,2,3,5]
}

df = pd.DataFrame(data)
df
```

```python
pivot_df = df.pivot_table(
            values = 'D',     # We want to aggregate the values of which column?
            index = 'A',      # We want to use which column as new index
            columns= 'C',     # We want to use the values of which column as the new columns?
            aggfunc= np.sum
)

pivot_df
```

```python
# Convert it back to a simple index

pivot_df.reset_index()
```

##### What is the mean temperature recorded by month?

```python
mean_temperature_df = weather_df.pivot_table(values = 'Temp (C)', index= weather_df['Date/Time'].dt.month, aggfunc= np.mean)
mean_temperature_df
```

#### Group By

```python
mean_temperature_df = weather_df.groupby(weather_df['Date/Time'].dt.month).agg("mean")
mean_temperature_df
```
